```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import pickle as pkl
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
card = pd.read_csv('fraudTrain.csv')
```

```python
card.head()
```

```python
card.shape
```

```python
card.isnull().sum()
```

```python
card.describe()
```

```python
X = card.drop('is_fraud', axis = 1)
```

```python
X.head()
```

```python
Y = card[['is_fraud']]
```

```python
Y.head()
```

```python
X.drop('trans_date_trans_time', axis = 1, inplace = True)
```

```python
X.head()
```

```python
X.drop('merchant', axis = 1, inplace = True)
```

```python
X.drop('category', axis = 1, inplace = True)
X.drop('first', axis = 1, inplace = True)
X.drop('last', axis = 1, inplace = True)
```

```python
X.head()
```

```python
X.drop('street', axis = 1, inplace = True)
X.drop('city', axis = 1, inplace = True)
X.drop('state', axis = 1, inplace = True)
X.drop('job', axis = 1, inplace = True)
```

```python
X.head()
```

```python
X.dtypes
```

```python
X['dob'] = X['dob'].str.replace('-','')
```

```python
X.head()
```

```python
X.dtypes
```

```python
X['gender'].replace({'M': 1 }, inplace = True)
X['gender'].replace({'F': 0 }, inplace = True)
```

```python
X.head()
```

```python
X.drop('trans_num', axis = 1, inplace = True)
```

```python
X1 = X.astype('float64')
```

```python
X1.head()
```

```python
X1.dtypes
```

```python
X1.corr()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, random_state = 2, test_size = 0.2)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train.shape
```

```python
Y_train.head()
```

```python
Y_train.shape
```

```python
Y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(Y_pred, Y_test)
```

```python
print(accuracy)
```
