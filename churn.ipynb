```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

Reading the CSV file

```python
churn = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
churn.head()
```

```python
churn['Geography']
```

```python
churn.corr()
```

```python
churnx = churn.drop('Exited', axis = 1)
```

```python
churnx.head()
```

```python
churnx = churnx.drop('RowNumber', axis = 1)
churnx = churnx.drop('CustomerId', axis = 1)
churnx = churnx.drop('Surname', axis = 1)
```

```python
churnx.head()
```

```python
churny = churn[['Exited']]
```

```python
churny.head()
```

```python
churnx = churnx.drop('Geography', axis = 1)
```

```python
churnx['CreditScore'] = churnx['CreditScore']/850
churnx['Age'] = churnx['Age']/92
churnx['Tenure'] = churnx['Tenure']/10
churnx['Balance'] = churnx['Balance']/250898.090000
churnx['EstimatedSalary'] = churnx['EstimatedSalary']/199992.480000
```

```python
churnx.head()
```

```python
churnx['Gender'].replace({'Male': 1 }, inplace = True)
```

```python
churnx.describe()
```

```python
churnx.tail()
```

```python
churnx['Gender'].replace({'Female': 0}, inplace = True)
```

```python
churnx.tail()
```

```python
churnx.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(churnx, churny, random_state = 2, stratify = churny, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, Y_train)
```

```python
churn_pred = classifier.predict(X_test)
accuracy = accuracy_score(churn_pred, Y_test)
```
